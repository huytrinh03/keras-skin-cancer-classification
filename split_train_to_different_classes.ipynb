```python
import pandas as pd
import os
import splitfolders as sf
```

```python
csv_path = "/Users/trinhnhathuy/pythonProject/ISIC-2019-venv/datasets/train/train-metadata.csv"
df = pd.read_csv(csv_path)
df
```

```python
# Group together rows whose value of "target" is 0 or 1
target_0 = df.loc[df['target'] == 0]
target_1 = df.loc[df['target'] == 1]
target_0
```

```python
# Take only the "isic_id" column
target_0_isic_id = target_0["isic_id"]
target_1_isic_id = target_1["isic_id"]
target_0_isic_id
```

```python
# Move to the correct directory
cd /Users/trinhnhathuy/pythonProject/ISIC-2019-venv/datasets/train
```

```python
# Sort image into the correct directory corresponding to their class
for element in target_0_isic_id:
    current_path = "{}.jpg".format(element)
    new_path = "target_0/{}.jpg".format(element)
    os.replace(current_path, new_path)
```

```python
# Sort image into the correct directory corresponding to their class
for element in target_1_isic_id:
    current_path = "{}.jpg".format(element)
    new_path = "target_1/{}.jpg".format(element)
    os.replace(current_path, new_path)
```

```python
# Use splitfolders package to split the train directory into test and validation subdirectory with ratio 80:20
input_folder = "/Users/trinhnhathuy/pythonProject/ISIC-2019-venv/datasets/train"
output_folder = "/Users/trinhnhathuy/pythonProject/ISIC-2019-venv/datasets/train_splitted"

sf.ratio(
    input_folder,
    output=output_folder,
    seed=1337,
    ratio=(0.8, 0.2),
    group_prefix=None,
    move=False)
```
